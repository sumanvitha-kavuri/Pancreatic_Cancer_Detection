```python
# Suppress warnings for clean output
import warnings
warnings.filterwarnings('ignore')

# Import libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from feature_engine.imputation import MeanMedianImputer
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_predict, StratifiedKFold
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Load the dataset
df = pd.read_csv('D:/Pancreatic/Debernardi et al 2020 data.csv')

# Handle missing values
numerical_cols = ['age', 'creatinine', 'LYVE1', 'REG1B', 'TFF1', 'REG1A']
imputer = MeanMedianImputer(imputation_method='median', variables=numerical_cols)
df[numerical_cols] = imputer.fit_transform(df[numerical_cols])

# Drop columns with excessive missing values
df = df.drop(['stage', 'benign_sample_diagnosis', 'plasma_CA19_9'], axis=1)

# Encode categorical variables
encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_cols = encoder.fit_transform(df[['sex', 'sample_origin']])
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out(['sex', 'sample_origin']), index=df.index)

# Combine features
df_processed = pd.concat([df[numerical_cols + ['diagnosis']], encoded_df], axis=1)

# Outlier removal with 1.5*IQR
for col in numerical_cols:
    Q1 = df_processed[col].quantile(0.25)
    Q3 = df_processed[col].quantile(0.75)
    IQR = Q3 - Q1
    df_processed = df_processed[~((df_processed[col] < (Q1 - 1.5 * IQR)) | (df_processed[col] > (Q3 + 1.5 * IQR)))]

# Scale numerical features
scaler = StandardScaler()
df_processed[numerical_cols] = scaler.fit_transform(df_processed[numerical_cols])

# Map diagnosis to binary classification (1,2=non-cancer, 3=cancer)
df_processed['diagnosis'] = df_processed['diagnosis'].map({1: 0, 2: 0, 3: 1})

# Apply SMOTE
X = df_processed.drop('diagnosis', axis=1)
y = df_processed['diagnosis']
smote = SMOTE(random_state=42)
X_balanced, y_balanced = smote.fit_resample(X, y)

# Split data
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.1, random_state=42)

# Define and train stacking model
xgb_model = XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss', n_estimators=1000, max_depth=11, learning_rate=0.02, subsample=0.95, colsample_bytree=0.8)
mlp_model = MLPClassifier(random_state=42, max_iter=3000, early_stopping=True, hidden_layer_sizes=(250, 150, 100), alpha=0.0001, learning_rate_init=0.001)
estimators = [('xgb', xgb_model), ('mlp', mlp_model)]
stacking_model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression(), cv=5)
stacking_model.fit(X_train, y_train)

# Predict and compute metrics
y_pred = stacking_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
class_report = classification_report(y_test, y_pred, output_dict=True)
cm = confusion_matrix(y_test, y_pred)

# Cross-validated accuracy
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_predictions = cross_val_predict(stacking_model, X_balanced, y_balanced, cv=skf, method='predict')
cv_accuracy = accuracy_score(y_balanced, cv_predictions)

# Create evaluation matrix
evaluation_matrix = {
    'Metric': ['Accuracy (Test Set)', 'Cross-Validated Accuracy', 
               'Precision (Non-Cancer)', 'Recall (Non-Cancer)', 'F1-Score (Non-Cancer)',
               'Precision (Cancer)', 'Recall (Cancer)', 'F1-Score (Cancer)'],
    'Value': [
        round(accuracy, 3),
        round(cv_accuracy, 3),
        round(class_report['0']['precision'], 3),
        round(class_report['0']['recall'], 3),
        round(class_report['0']['f1-score'], 3),
        round(class_report['1']['precision'], 3),
        round(class_report['1']['recall'], 3),
        round(class_report['1']['f1-score'], 3)
    ]
}

# Display evaluation matrix
eval_df = pd.DataFrame(evaluation_matrix)
print("\nEvaluation Matrix:")
print(eval_df.to_string(index=False))

# Display confusion matrix
print("\nConfusion Matrix:")
print("[[True Negatives, False Positives]")
print(" [False Negatives, True Positives]]")
print(cm)
```
